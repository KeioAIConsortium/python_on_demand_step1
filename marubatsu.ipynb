## 第3回：手が置けるか判定しよう

**習得する文法・概念**：`if` / `elif` / `else`、比較演算子（`==`, `!=`, `<`, `>`）、論理演算子（`and`, `or`, `not`）

````python
# 第3回：制御文（if/elif/else）

board = [" "] * 9
current_player = "X"

# ---- 手の有効性判定 ----
position = 4

if position < 0 or position > 8:
    print("範囲外のマスです")
elif board[position] != " ":
    print("すでに置かれているマスです")
else:
    board[position] = current_player
    print(f"マス{position}に{current_player}を置きました")

print("盤面:", board)

# ---- 手番の切り替え ----
print("現在の手番:", current_player)

if current_player == "X":
    current_player = "O"
else:
    current_player = "X"

print("切り替え後の手番:", current_player)

# ---- すでに埋まったマスへの試み ----
position = 4  # さっき置いたマス

if position < 0 or position > 8:
    print("範囲外のマスです")
elif board[position] != " ":
    print("すでに置かれているマスです")
else:
    board[position] = current_player
    print(f"マス{position}に{current_player}を置きました")
````

## 第4回：勝敗を判定しよう
**木構造上の現在地**：GUI付きまるバツゲーム > ゲームロジック > **勝敗を判定する**  
**習得する文法・概念**：`for x in list`、リストのリスト、`range()`、`None`の扱い

````python
# 第4回：forループ

# ---- 勝利パターン（8通り）をリストのリストで定義 ----
win_patterns = [
    [0, 1, 2],  # 上段・横
    [3, 4, 5],  # 中段・横
    [6, 7, 8],  # 下段・横
    [0, 3, 6],  # 左列・縦
    [1, 4, 7],  # 中列・縦
    [2, 5, 8],  # 右列・縦
    [0, 4, 8],  # 左上→右下・斜め
    [2, 4, 6],  # 右上→左下・斜め
]

# ---- 勝者チェック ----
board = ["X", "X", "X",
         " ", "O", " ",
         "O", " ", " "]

winner = None
for a, b, c in win_patterns:
    if board[a] == board[b] == board[c] and board[a] != " ":
        winner = board[a]
        break

print("勝者:", winner)

# ---- 引き分けチェック ----
draw_board = ["X", "O", "X",
              "X", "O", "O",
              "O", "X", "X"]

winner = None
for a, b, c in win_patterns:
    if draw_board[a] == draw_board[b] == draw_board[c] and draw_board[a] != " ":
        winner = draw_board[a]
        break

if winner:
    print("勝者:", winner)
elif " " not in draw_board:
    print("引き分け！")
else:
    print("ゲーム継続中")
````

## 第5回：コードを部品に整理しよう
**木構造上の現在地**：GUI付きまるバツゲーム > ゲームロジック > **コードを部品に整理する**  
**習得する文法・概念**：`def`、引数、`return`、関数呼び出し、コードの整理・再利用

> 第1〜4回のフラットコードを関数にリファクタリングする

````python
# 第5回：関数（第1〜4回のフラットコードをリファクタリング）

def initialize_game():
    """盤面と手番を初期化して返す"""
    board = [" "] * 9
    current_player = "X"
    return board, current_player

def display_board(board):
    """盤面を3×3で表示する"""
    print(f" {board[0]} | {board[1]} | {board[2]} ")
    print("-----------")
    print(f" {board[3]} | {board[4]} | {board[5]} ")
    print("-----------")
    print(f" {board[6]} | {board[7]} | {board[8]} ")

def is_valid_move(board, position):
    """指定位置に置けるかどうかを返す"""
    if position < 0 or position > 8:
        return False
    if board[position] != " ":
        return False
    return True

def place_mark(board, position, player):
    """有効な手なら盤面にマークを置き、成否を返す"""
    if is_valid_move(board, position):
        board[position] = player
        return True
    return False

def switch_player(current_player):
    """手番を切り替えて返す"""
    if current_player == "X":
        return "O"
    else:
        return "X"

def check_winner(board):
    """勝者がいれば 'X' または 'O'、いなければ None を返す"""
    win_patterns = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],
        [0, 3, 6], [1, 4, 7], [2, 5, 8],
        [0, 4, 8], [2, 4, 6],
    ]
    for a, b, c in win_patterns:
        if board[a] == board[b] == board[c] and board[a] != " ":
            return board[a]
    return None

def check_draw(board):
    """引き分けなら True を返す"""
    return " " not in board

def get_game_status(board):
    """ゲーム状態を文字列で返す。継続中は None を返す"""
    winner = check_winner(board)
    if winner:
        return f"{winner} の勝ち！"
    if check_draw(board):
        return "引き分け！"
    return None

# ---- 動作確認 ----
board, current_player = initialize_game()
place_mark(board, 0, "X")
place_mark(board, 1, "X")
place_mark(board, 2, "X")
display_board(board)
print("ゲーム状態:", get_game_status(board))
````

## 第6回：ゲームを繰り返し進行しよう
**木構造上の現在地**：GUI付きまるバツゲーム > ゲームロジック > **ゲームを進行する**  
**習得する文法・概念**：`while True`、`break`、`continue`、`input()`、`int()`変換、`try`/`except`（軽く）

````python
# 第6回：whileループ

def play_game():
    """1ゲーム分のCUI対戦を実行する"""
    board, current_player = initialize_game()

    while True:
        display_board(board)
        print(f"\n【{current_player} の番】0〜8 の番号を入力してください。")

        try:
            position = int(input("位置を入力: "))
        except ValueError:
            print("数字を入力してください。")
            continue

        if not place_mark(board, position, current_player):
            print("そのマスには置けません。別のマスを選んでください。")
            continue

        status = get_game_status(board)
        if status:
            display_board(board)
            print(status)
            break

        current_player = switch_player(current_player)

# ※ このセルを実行するとCUI版まるバツゲームで遊べます
# play_game()
````

## 第7回：CUI版ゲームを仕上げよう
**木構造上の現在地**：GUI付きまるバツゲーム > ゲームロジック > **CUI版ゲーム完成**  
**習得する文法・概念**：大きな新文法なし。関数の再利用、全体統合の考え方

````python
# 第7回：統合・仕上げ（リプレイ機能追加）

def play_game_loop():
    """リプレイ機能付きCUI版まるバツゲームを実行する"""
    while True:
        play_game()
        again = input("\nもう一度遊びますか？ (y/n): ")
        if again.lower() != "y":
            print("ありがとうございました！")
            break

# ※ このセルを実行するとリプレイ機能付きCUI版で遊べます
# play_game_loop()
````

## 第8回：Gradioで画面を作ろう
**木構造上の現在地**：GUI付きまるバツゲーム > Gradio UI > **画面を作る**  
**習得する文法・概念**：`import`、ライブラリの使い方、Gradioコンポーネント（`gr.Blocks`, `gr.Button`, `gr.Row`, `gr.Textbox`）

````python
# Gradio のインストール（Google Colab では毎回実行が必要）
!pip install gradio -q
````

````python
# 第8回：import・Gradio基礎（UI骨組みの作成）

import gradio as gr

# ---- CSSでUI全体を装飾する ----
CSS = """
/* マスボタン：大きな正方形・グリッド風ボーダー */
.cell-btn {
    min-height: 110px !important;
}
.cell-btn button, .cell-btn button:not(:disabled) {
    font-size: 2.5rem !important;
    font-weight: bold !important;
    min-height: 110px !important;
    border: 3px solid #aaa !important;
    border-radius: 6px !important;
    background: #fafafa !important;
    color: #222 !important;
}
.cell-btn button:hover {
    background: #e8f0ff !important;
    border-color: #5588ff !important;
}
/* ステータス欄：中央揃え・大きめフォント */
.status-box label { font-weight: bold; }
.status-box textarea {
    text-align: center !important;
    font-size: 1.2rem !important;
    font-weight: bold !important;
}
/* リセットボタン行を中央に */
.reset-row {
    justify-content: center !important;
    max-width: 240px !important;
    margin: 4px auto !important;
}
"""

def build_ui_skeleton():
    """ボタン9個・ステータス表示・リセットボタンのUI骨組みを作る"""
    with gr.Blocks(title="まるバツゲーム", css=CSS) as app:

        # タイトル
        gr.Markdown(
            "# ⭕❌ まるバツゲーム\n"
            "<p style='text-align:center; color:#666;'>"
            "先手は X です。マスをクリックして手を打ちましょう。</p>"
        )

        # ステータスメッセージ
        status_msg = gr.Textbox(
            value="X の番です",
            label="ゲームの状態",
            interactive=False,
            elem_classes=["status-box"]
        )

        # 3×3 ボタングリッド
        # ポイント：gr.Row() を直接 Blocks の直下に置き、
        #           各ボタンに scale=1 を指定して等幅にする
        buttons = []
        for row in range(3):
            with gr.Row():
                for col in range(3):
                    btn = gr.Button(
                        " ",
                        size="lg",
                        scale=1,              # ← 3列を均等幅にする
                        elem_classes=["cell-btn"]
                    )
                    buttons.append(btn)

        # リセットボタン（中央寄せ）
        with gr.Row(elem_classes=["reset-row"]):
            reset_btn = gr.Button("🔄 リセット", variant="secondary", scale=0)

    return app

# ※ このセルを実行するとUI骨組みが表示されます
app = build_ui_skeleton()
# app.launch()
````

## 第9回：GUIとロジックをつないで完成させよう
**木構造上の現在地**：GUI付きまるバツゲーム > Gradio UI > **完成**  
**習得する文法・概念**：イベント駆動プログラミング、コールバック、`gr.State`、関数とUIの接続

````python
# 第9回：イベント処理（クリック処理・リセット接続、GUI版完成）

import gradio as gr

# CSS は第8回で定義済みの変数をそのまま使用

def build_ui():
    with gr.Blocks(title="まるバツゲーム", css=CSS) as app:  # ← CSS を適用

        gr.Markdown(
            "# ⭕❌ まるバツゲーム\n"
            "<p style='text-align:center; color:#666;'>"
            "先手は X です。マスをクリックして手を打ちましょう。</p>"
        )

        # --- 状態の保持 ---
        board_state  = gr.State([" "] * 9)
        player_state = gr.State("X")

        # --- 状態メッセージ ---
        status_msg = gr.Textbox(
            value="X の番です",
            label="ゲームの状態",
            interactive=False,
            elem_classes=["status-box"]   # ← 中央揃え・太字
        )

        # --- 3×3 ボタングリッド ---
        buttons = []
        for row in range(3):
            with gr.Row():
                for col in range(3):
                    btn = gr.Button(
                        " ",
                        size="lg",
                        scale=1,                       # ← 3列を均等幅に
                        elem_classes=["cell-btn"]      # ← 大きな正方形ボタン
                    )
                    buttons.append(btn)

        # --- リセットボタン（中央寄せ）---
        with gr.Row(elem_classes=["reset-row"]):
            reset_btn = gr.Button("🔄 リセット", variant="secondary", scale=0)

        # 出力先：board_state, player_state, ボタン×9, status_msg
        outputs = [board_state, player_state] + buttons + [status_msg]

        def _board_to_labels(board):
            return [cell for cell in board]

        def make_click_handler(position):
            """各ボタン用のクリックハンドラを生成する"""
            def click_handler(board, player):
                if get_game_status(board) or not is_valid_move(board, position):
                    labels = _board_to_labels(board)
                    status = get_game_status(board)
                    msg = status if status else f"{player} の番です"
                    return [board, player] + labels + [msg]

                place_mark(board, position, player)

                status = get_game_status(board)
                if status:
                    next_player = player
                    msg = status
                else:
                    next_player = switch_player(player)
                    msg = f"{next_player} の番です"

                labels = _board_to_labels(board)
                return [board, next_player] + labels + [msg]

            return click_handler

        def reset_handler():
            board, player = initialize_game()
            labels = _board_to_labels(board)
            return [board, player] + labels + ["X の番です"]

        # --- クリックイベントを各ボタンに接続 ---
        for i, btn in enumerate(buttons):
            btn.click(
                fn=make_click_handler(i),
                inputs=[board_state, player_state],
                outputs=outputs
            )

        # --- リセットボタンを接続 ---
        reset_btn.click(
            fn=reset_handler,
            inputs=[],
            outputs=outputs
        )

    return app
````

````python
# 第9回：アプリを起動して完成！

app = build_ui()
app.launch()
````
