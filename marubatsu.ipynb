## 第1回：まるバツゲームの盤面を作ろう
**木構造上の現在地**：GUI付きまるバツゲーム > ゲームロジック > **盤面を表現する**  
**習得する文法・概念**：値、変数、リスト、(print関数)

関数は第2回で教える

````python
# 第1回：盤面を表現する

def initialize_game():
    board = [" "] * 9   # 0〜8 のインデックスで9マスを管理
    current_player = "X"
    return board, current_player

# 動作確認
board, current_player = initialize_game()
print("盤面データ:", board)
print("最初の手番:", current_player)
````

## 第2回：盤面を見やすく表示しよう
**木構造上の現在地**：GUI付きまるバツゲーム > ゲームロジック > **盤面を表示する**  
**習得する文法・概念**：関数定義、引数、文字列の連結、`print`

````python
# 第2回：盤面を表示する

def display_board(board):
    row1 = f" {board[0]} | {board[1]} | {board[2]} "
    row2 = f" {board[3]} | {board[4]} | {board[5]} "
    row3 = f" {board[6]} | {board[7]} | {board[8]} "
    separator = "-----------"
    print(row1)
    print(separator)
    print(row2)
    print(separator)
    print(row3)

# 動作確認
board, current_player = initialize_game()
display_board(board)
````

## 第3回：プレイヤーの手を受け付けよう
**木構造上の現在地**：GUI付きまるバツゲーム > ゲームロジック > **手を受け付ける**  
**習得する文法・概念**：条件分岐、比較演算子、`if` / `elif` / `else`

````python
# 第3回：手を受け付ける

def is_valid_move(board, position):
    if position < 0 or position > 8:
        return False
    if board[position] != " ":
        return False
    return True

def place_mark(board, position, player):
    if is_valid_move(board, position):
        board[position] = player
        return True
    return False

# 動作確認
board, current_player = initialize_game()
print("マス4に置く前:", board)
place_mark(board, 4, "X")
print("マス4に置いた後:", board)
print("すでに埋まったマスに置こうとすると:", is_valid_move(board, 4))
````

## 第4回：手番を切り替えながら1手ずつ進めよう
**木構造上の現在地**：GUI付きまるバツゲーム > ゲームロジック > **手番を切り替える**  
**習得する文法・概念**：代入、再代入、状態管理、条件式を使った切り替え

````python
# 第4回：手番を切り替える

def switch_player(current_player):
    if current_player == "X":
        return "O"
    else:
        return "X"

# 動作確認
board, current_player = initialize_game()
print("最初の手番:", current_player)

place_mark(board, 4, current_player)
current_player = switch_player(current_player)
print("1手打った後の手番:", current_player)

place_mark(board, 0, current_player)
current_player = switch_player(current_player)
print("2手打った後の手番:", current_player)

display_board(board)
````

## 第5回：勝ったかどうかを判定しよう
**木構造上の現在地**：GUI付きまるバツゲーム > ゲームロジック > **勝敗を判定する**  
**習得する文法・概念**：複数条件の比較、論理演算、リストによる組み合わせ管理

````python
# 第5回：勝敗を判定する

def check_winner(board):
    win_patterns = [
        [0, 1, 2],  # 上段・横
        [3, 4, 5],  # 中段・横
        [6, 7, 8],  # 下段・横
        [0, 3, 6],  # 左列・縦
        [1, 4, 7],  # 中列・縦
        [2, 5, 8],  # 右列・縦
        [0, 4, 8],  # 左上→右下・斜め
        [2, 4, 6],  # 右上→左下・斜め
    ]
    for a, b, c in win_patterns:
        if board[a] == board[b] == board[c] and board[a] != " ":
            return board[a]
    return None

# 動作確認
board, _ = initialize_game()
board[0] = "X"
board[1] = "X"
board[2] = "X"
display_board(board)
print("勝者:", check_winner(board))
````

## 第6回：引き分けとゲーム終了を判定しよう
**木構造上の現在地**：GUI付きまるバツゲーム > ゲームロジック > **引き分けを判定する**  
**習得する文法・概念**：否定条件、`None` の扱い、複数条件の整理

````python
# 第6回：引き分けを判定する

def check_draw(board):
    return " " not in board

def get_game_status(board):
    winner = check_winner(board)
    if winner:
        return f"{winner} の勝ち！"
    if check_draw(board):
        return "引き分け！"
    return None   # None = ゲーム継続中

# 動作確認（引き分けの盤面）
draw_board = ["X", "O", "X",
              "X", "O", "O",
              "O", "X", "X"]
display_board(draw_board)
print("ゲーム状態:", get_game_status(draw_board))
````

## 第7回：ゲーム全体を動かす処理をまとめよう
**木構造上の現在地**：GUI付きまるバツゲーム > ゲームロジック > **ゲーム全体を進行する**  
**習得する文法・概念**：関数同士の連携、処理の順序設計、繰り返し処理

````python
# 第7回：ゲーム全体を進行する

def play_game():
    board, current_player = initialize_game()

    while True:
        display_board(board)
        print(f"\n【{current_player} の番】0〜8 の番号を入力してください。")

        try:
            position = int(input("位置を入力: "))
        except ValueError:
            print("数字を入力してください。")
            continue

        if not place_mark(board, position, current_player):
            print("そのマスには置けません。別のマスを選んでください。")
            continue

        status = get_game_status(board)
        if status:
            display_board(board)
            print(status)
            break

        current_player = switch_player(current_player)

# ※ このセルを実行するとCUI版まるバツゲームで遊べます
# play_game()
````

## 第8回：もう一度遊べるように初期化しよう
**木構造上の現在地**：GUI付きまるバツゲーム > ゲームロジック > **ゲームを初期化する**  
**習得する文法・概念**：初期値、関数による状態の再設定、処理の切り出し・再利用

````python
# 第8回：ゲームを初期化する
# ※ initialize_game() は第1回で定義済み。
#    ここでは「再プレイのためにリセットして繰り返す」ループを完成させる。

def play_game_loop():
    while True:
        play_game()
        again = input("\nもう一度遊びますか？ (y/n): ")
        if again.lower() != "y":
            print("ありがとうございました！")
            break

# ※ このセルを実行するとリセット機能付きCUI版で遊べます
# play_game_loop()
````

## 第9回：Gradioで画面を作ろう
**木構造上の現在地**：GUI付きまるバツゲーム > Gradio UI > **ボタンを配置する / 状態メッセージを表示する**  
**習得する文法・概念**：`import`、ライブラリ利用の基本、コンポーネント、GUIとデータの対応付け

````python
# Gradio のインストール（Google Colab では毎回実行が必要）
!pip install gradio -q
````

````python
# 第9回：Gradio UIの骨組みを作る（ボタン配置・状態メッセージ）

import gradio as gr

def build_ui():
    with gr.Blocks(title="まるバツゲーム") as app:

        gr.Markdown("# ⭕❌ まるバツゲーム")

        # --- 状態の保持 ---
        board_state  = gr.State([" "] * 9)
        player_state = gr.State("X")

        # --- 状態メッセージ（第9回で配置）---
        status_msg = gr.Textbox(
            value="X の番です",
            label="ゲームの状態",
            interactive=False
        )

        # --- 3×3 ボタングリッド（第9回で配置）---
        buttons = []
        for row in range(3):
            with gr.Row():
                for col in range(3):
                    btn = gr.Button(" ", size="lg", min_width=80)
                    buttons.append(btn)

        # --- リセットボタン（第9回で配置）---
        reset_btn = gr.Button("🔄 リセット", variant="secondary")

        # =====================================================
        # 第10回：クリック処理・盤面反映・リセット接続
        # =====================================================

        # 出力先：board_state, player_state, ボタン×9, status_msg
        outputs = [board_state, player_state] + buttons + [status_msg]

        def _board_to_labels(board):
            """盤面リストをボタン用ラベルのリストに変換する"""
            return [cell for cell in board]

        def make_click_handler(position):
            """各ボタン用のクリックハンドラを生成する"""
            def click_handler(board, player):
                # ゲーム終了済み or 置けないマスは何もしない
                if get_game_status(board) or not is_valid_move(board, position):
                    labels = _board_to_labels(board)
                    status = get_game_status(board)
                    msg = status if status else f"{player} の番です"
                    return [board, player] + labels + [msg]

                # マークを置く
                place_mark(board, position, player)

                # ゲーム状態を確認
                status = get_game_status(board)
                if status:
                    next_player = player
                    msg = status
                else:
                    next_player = switch_player(player)
                    msg = f"{next_player} の番です"

                labels = _board_to_labels(board)
                return [board, next_player] + labels + [msg]

            return click_handler

        def reset_handler():
            board, player = initialize_game()
            labels = _board_to_labels(board)
            return [board, player] + labels + ["X の番です"]

        # --- クリックイベントを各ボタンに接続 ---
        for i, btn in enumerate(buttons):
            btn.click(
                fn=make_click_handler(i),
                inputs=[board_state, player_state],
                outputs=outputs
            )

        # --- リセットボタンを接続 ---
        reset_btn.click(
            fn=reset_handler,
            inputs=[],
            outputs=outputs
        )

    return app
````
